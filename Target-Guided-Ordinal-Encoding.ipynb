```python
import pandas as pd
```

```python
df = pd.DataFrame({'city':['jaipur','churu','jhunjhunu','chomu','chomu','jaipur'],
                  'price':[200,123,543,789,456,651]})
```

```python
df
```

```python
#Calculate the mean price 
```

```python
g = df.groupby('city')
```

```python
g.mean()
```

```python
mean = df.groupby('city')['price'].mean().to_dict()
```

```python
#replace each city with mean 
df['city_mean'] = df['city'].map(mean)
```
