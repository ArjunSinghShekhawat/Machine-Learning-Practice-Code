```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
obj = MinMaxScaler()
```

```python
pd.DataFrame(obj.fit_transform(df[['passengers','distance','fare','tip','tolls','total']]),columns=obj.get_feature_names_out())
```
